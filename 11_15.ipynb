###1.구글 드라이브에서 사진 접근권한 가져오기

```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
```

###2.cv2.imread로 사진을 불러온다

```python
src = cv2.imread('/content/drive/MyDrive/my_pet.jpg', cv2.IMREAD_COLOR)
```

###3.이미지를 흑백으로 전환시켜 gray에 저장한다.

```python
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
```

###4.gary는 sobel필터가 적용될 대상
###cv2.CV_8U 출력 이미지의 데이터 타입을 나타냅니다.
###1, 0 : Sobel 커널의 x-방향 미분 차수입니다.
###3 : Sobel 커널의 크기입니다.

```python
sobel = cv2.Sobel(gray, cv2.CV_8U, 1, 0, 3)
```

###5.cv2_imshow(sobel) sobel을 이미지로 출력합니다.

```python
cv2_imshow(sobel)
```
